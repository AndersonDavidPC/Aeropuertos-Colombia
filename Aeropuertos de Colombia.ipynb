```python
import pandas as pd
```

```python
df = pd.read_json('https://api-colombia.com/api/v1/airport')
df.head()
```

```python
df_militar = df[df['type'] == 'Militar']
df_militar.head()
```

```python
df_nal = df[df['type'] == 'Nacional']
df_nal.head()
```

```python
df_int = df[df['type'] == 'Internacional']
df_int.head()
```

```python
import geopandas as gpd
```

```python
import matplotlib.pyplot as plt
```

```python
# Crear una figura y un eje
fig, ax = plt.subplots(figsize=(10, 10))

# Crear un GeoDataFrame a partir del DataFrame filtrado
gdf_militar = gpd.GeoDataFrame(df_militar, geometry=gpd.points_from_xy(df_militar['longitude'], df_militar['latitude']))

# Representar los aeropuertos en el plano
gdf_militar.plot(ax=ax, color='green', markersize=10)

# Configurar el aspecto del plano
ax.set_title('Posición geográfica de los aeropuertos militares')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')

# Mostrar el plano
plt.show()
```
